```python
def get_ft(data, h, T):
    Fs = 1/h #sampling rate
    n = int(T/h) #number of observations
    
    df_fft = pd.DataFrame()
    df_fft["fft"] = fftpack.fft(data)
    df_fft["A"] = df_fft["fft"].abs()
    df_fft["fr"] = Fs/n * np.linspace(0,n,int(n))
    return(df_fft)
```

```python
ft = get_ft(f3, 0.01, int(6*np.pi))
```
