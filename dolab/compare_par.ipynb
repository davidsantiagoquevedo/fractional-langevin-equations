```python
import pandas as pd
import sys
sys.path.append("../src/")
import utils as ut
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
def compare(H, A, eta, C, theta_12, theta_H, file_100 = False):
    h = 0.005
    T = 20
    n = int(T/h)
    t = np.arange(0, T, h)
    
    data_path = "../data/two_baths/"
    file = f"par_msd-h-{H}-A{A}-eta{eta}-C{C}-t12_{theta_12}-tH{theta_H}.hdf5"
    df1 = ut.read_hdf5_data(file)

    file = f"msd-h-{H}-A{A}-eta{eta}-C{C}-t12_{theta_12}-tH{theta_H}.hdf5"
    df2 = ut.read_hdf5_data(file)

    fig, ax = plt.subplots()
    ax.plot(t, df1.msd, label = "Serial")
    ax.plot(t, df2.msd, label = "Parallel")
    if file_100: 
        h = 0.1
        file = f"par_dt{h}-T{T}_msd-h-{H}-A{A}-eta{eta}-C{C}-t12_{theta_12}-tH{theta_H}.hdf5"
        df3 = ut.read_hdf5_data(data_path + file) 
        ax.plot(df3.t, df3.msd, label = "Parallel")
        
    ax.set_title(f"H = {H}")
    ax.legend()
```

```python
A = 1.0; eta = 1.0; C = 0.0; theta_H = 0.0; theta_12 = 1.0; H = 0.95
compare(H, A, eta, C, theta_12, theta_H, file_100 = True)
```
